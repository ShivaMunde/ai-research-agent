```python
from ddgs import DDGS
from groq import Groq

API_KEY = input("Enter your Groq API Key: ")

client = Groq(api_key=API_KEY)
```

```python
def search_web(query, num_results=5):

    results = []

    with DDGS() as ddgs:
        for r in ddgs.text(query, max_results=num_results):
            results.append(r["body"])

    return "\n".join(results)
```

```python
def summarize_with_ai(text):

    completion = client.chat.completions.create(
        model="llama-3.1-8b-instant",
        messages=[
            {
                "role": "system",
                "content": "You are a helpful research assistant."
            },
            {
                "role": "user",
                "content": f"Summarize the following research findings into clear bullet points:\n\n{text}"
            }
        ]
    )

    return completion.choices[0].message.content
```

```python
def research_agent(query):

    print("Searching the web...\n")

    search_results = search_web(query)

    print("Search results collected.\n")

    print("Generating AI summary...\n")

    summary = summarize_with_ai(search_results)

    return summary
```

```python
query = input("Enter research topic: ")

result = research_agent(query)

print("\n==============================")
print("AI RESEARCH SUMMARY")
print("==============================\n")

print(result)
```
